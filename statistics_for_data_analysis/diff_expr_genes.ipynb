```python
import pandas as pd
data = pd.read_csv('gene.csv')
```

```python
normal_data = data[data['Diagnosis'] == 'normal']
```

```python
normal_data
```

```python
data
```

```python
early_data = data[data['Diagnosis'] == 'early neoplasia']
cancer_data = data[data['Diagnosis'] == 'cancer']
```

```python
columns = data.columns.tolist()
columns[1]
```

```python
from scipy.stats import ttest_ind
ttest_ind(normal_data[columns[2]], early_data[columns[2]], equal_var = False).pvalue
```

```python
s = []
for i in range (2, len(columns)):
    s.append (ttest_ind(normal_data[columns[i]], early_data[columns[i]], equal_var = False).pvalue)
s
```

```python
ans1 = sum(1 for i in s if i < 0.05)
```

```python
ans1
```

```python
import statsmodels.stats.multitest as smm
reject, p_corrected, a1, a2 = smm.multipletests(s, alpha = 0.025, method = 'fdr_bh')
```

```python
sum (1 for i in reject if i == True)
```

```python
def change_fold(C, T):
    if C > T:
        return float(-C)/T
    else:
        return float(T)/C
```

```python
fold = []
for i in range (len(reject)):
    if reject[i] == True:
        fold.append(change_fold(normal_data[columns[i+2]].mean(), early_data[columns[i+2]].mean()))
```

```python
fold
```

```python
s = []
for i in range (2, len(columns)):
    s.append (ttest_ind(cancer_data[columns[i]], early_data[columns[i]], equal_var = False).pvalue)
ans2 = sum(1 for i in s if i < 0.05)
reject, p_corrected, a1, a2 = smm.multipletests(s, alpha = 0.025, method = 'fdr_bh')
fold = []
for i in range (len(reject)):
    if reject[i] == True:
        fold.append(change_fold(cancer_data[columns[i+2]].mean(), early_data[columns[i+2]].mean()))
fold
```

```python
ans4 = sum(1 for i in fold if abs(i) > 1.5)
```
